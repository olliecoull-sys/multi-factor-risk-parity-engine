```python
def fetch_factor_proxies(tickers, start_date, end_date):
    """
    Downloads adjusted closing prices from Yahoo Finance and calculates log returns.
    """
    print(f"Fetching historical EOD data for: {', '.join(tickers)}")
    try:
        data = yf.download(tickers, start=start_date, end=end_date, progress=False)
        if 'Adj Close' in data.columns:
            prices = data['Adj Close']
        else:
            prices = data['Close']

        # Align tickers accurately
        prices = prices[tickers]

        # Clean data gracefully
        if prices.isnull().values.any():
            prices = prices.ffill().bfill()

        returns = np.log(prices / prices.shift(1)).dropna()
        return prices, returns
    except Exception as e:
        raise RuntimeError(f"Data pipeline failure: {e}")

# Define asset-factor mapping configuration
ticker_config = {
    'SPY': 'Market Beta',
    'IWM': 'Size',
    'VTV': 'Value',
    'MTUM': 'Momentum',
    'QUAL': 'Quality'
}
tickers = list(ticker_config.keys())
start_date = "2015-01-01"
end_date = datetime.today().strftime('%Y-%m-%d')

prices, returns = fetch_factor_proxies(tickers, start_date, end_date)
print(f"Data shape: {returns.shape}. Range: {returns.index[0].date()} to {returns.index[-1].date()}")
returns.rename(columns=ticker_config).head()
```

```python
def calculate_shrinkage_covariance(returns, annualization_factor=252):
    """
    Applies Ledoit-Wolf shrinkage to the empirical covariance matrix.
    """
    lw = LedoitWolf()
    lw.fit(returns.values)
    shrinked_cov = lw.covariance_ * annualization_factor

    # Reconstruct DataFrame with matching metadata
    cov_df = pd.DataFrame(shrinked_cov, index=returns.columns, columns=returns.columns)
    return cov_df

annualized_returns = returns.mean() * 252
shrinked_covariance = calculate_shrinkage_covariance(returns)

fig, ax = plt.subplots(1, 2, figsize=(14, 5))
sns.heatmap(returns.cov() * 252, annot=True, cmap="coolwarm", fmt=".4f", ax=ax[0])
ax[0].set_title("Sample Covariance Matrix")
sns.heatmap(shrinked_covariance, annot=True, cmap="coolwarm", fmt=".4f", ax=ax[1])
ax[1].set_title("Ledoit-Wolf Shrinkage Covariance Matrix")
plt.tight_layout()
plt.show()
```

```python
def compute_quasi_diagonal(link_matrix):
    """
    Sorts the cluster link matrix recursively to structure similar components closely.
    """
    link_matrix = link_matrix.astype(int)
    sort_idx = pd.Series([link_matrix[-1, 0], link_matrix[-1, 1]])
    num_items = link_matrix[-1, 3]

    while sort_idx.max() >= num_items:
        sort_idx.index = range(0, sort_idx.shape[0] * 2, 2)
        df0 = sort_idx[sort_idx >= num_items]
        i = df0.index
        j = df0.values - num_items
        sort_idx[i] = link_matrix[j, 0]
        df0 = pd.Series(link_matrix[j, 1], index=i + 1)
        sort_idx = pd.concat([sort_idx, df0])
        sort_idx = sort_idx.sort_index()
        sort_idx.index = range(sort_idx.shape[0])
    return sort_idx.tolist()

def compute_cluster_variance(cov, cluster_items):
    """
    Calculates the variance of a specific cluster allocation.
    """
    sub_cov = cov.loc[cluster_items, cluster_items]
    inv_diag = 1.0 / np.diag(sub_cov)
    weights = inv_diag / np.sum(inv_diag)
    return np.dot(weights, np.dot(sub_cov, weights))

def recursive_bisection(cov, sort_items):
    """
    Performs recursive risk allocation through top-down bisection partitioning.
    """
    weights = pd.Series(1.0, index=sort_items)
    queue = [sort_items]

    while len(queue) > 0:
        items = queue.pop(0)
        if len(items) <= 1:
            continue

        mid = len(items) // 2
        left_cluster = items[:mid]
        right_cluster = items[mid:]

        var_l = compute_cluster_variance(cov, left_cluster)
        var_r = compute_cluster_variance(cov, right_cluster)
        alpha = 1.0 - var_l / (var_l + var_r)

        weights[left_cluster] *= alpha
        weights[right_cluster] *= (1.0 - alpha)

        queue.append(left_cluster)
        queue.append(right_cluster)
    return weights

# Generate distance metrics based on factor return correlations
correlation = returns.corr()
distance_matrix = np.sqrt(2 * (1 - correlation).clip(lower=0))
links = linkage(distance_matrix, method='single')

# Generate risk parity weights
ordered_indices = compute_quasi_diagonal(links)
sorted_tickers = [returns.columns[i] for i in ordered_indices]
hrp_weights = recursive_bisection(shrinked_covariance, sorted_tickers)
hrp_weights = hrp_weights.reindex(tickers)

print("Optimal Factor Proxy Allocation Weights:")
for tick, wt in zip(tickers, hrp_weights):
    print(f"  {ticker_config[tick]:<12} ({tick}): {wt*100:6.2f}%")
```

```python
def run_rolling_backtest(returns, ticker_config, window_size=252):
    """
    Executes an out-of-sample backtest rebalancing weights systematically every 21 days.
    """
    n_obs = len(returns)
    portfolio_returns = []
    dates = []
    current_weights = None

    # Core out-of-sample looping iteration
    for i in range(window_size, n_obs):
        # Regular rebalancing interval checkpoint
        if (i - window_size) % 21 == 0:
            historical_window = returns.iloc[i-window_size:i]
            cov_matrix = calculate_shrinkage_covariance(historical_window)
            corr = historical_window.corr()
            dist = np.sqrt(2 * (1 - corr).clip(lower=0))
            link_mat = linkage(dist, method='single')
            sorted_idx = compute_quasi_diagonal(link_mat)
            sorted_tk = [historical_window.columns[j] for j in sorted_idx]

            w_optimized = recursive_bisection(cov_matrix, sorted_tk)
            current_weights = w_optimized.reindex(returns.columns).values

        day_return = np.dot(returns.iloc[i].values, current_weights)
        portfolio_returns.append(day_return)
        dates.append(returns.index[i])

    return pd.Series(portfolio_returns, index=dates)

portfolio_perf = run_rolling_backtest(returns, ticker_config)
equal_weight_perf = returns.iloc[252:].mean(axis=1)
spy_benchmark = returns['SPY'].iloc[252:]

print("Backtest execution loop completed successfully.")
```

```python
def generate_risk_analytics(perf_series, name):
    """
    Computes performance summary data.
    """
    ann_ret = np.exp(perf_series.mean() * 252) - 1
    ann_vol = perf_series.std() * np.sqrt(252)
    sharpe = (perf_series.mean() * 252) / (perf_series.std() * np.sqrt(252)) # Assumes R_f = 0

    cum_ret = np.exp(perf_series.cumsum())
    peak = cum_ret.cummax()
    drawdown = (cum_ret - peak) / peak
    max_dd = drawdown.min()

    return {
        'Annualized Return': f"{ann_ret*100:.2f}%",
        'Annualized Volatility': f"{ann_vol*100:.2f}%",
        'Sharpe Ratio': f"{sharpe:.2f}",
        'Max Drawdown': f"{max_dd*100:.2f}%"
    }

metrics_df = pd.DataFrame({
    'HRP Factor Engine': generate_risk_analytics(portfolio_perf, 'HRP'),
    'Equal Weight': generate_risk_analytics(equal_weight_perf, 'EW'),
    'S&P 500 (SPY)': generate_risk_analytics(spy_benchmark, 'SPY')
})

# Render Dashboard
plt.figure(figsize=(12, 6))
plt.plot(np.exp(portfolio_perf.cumsum()) * 100, label='HRP Factor Engine', linewidth=2, color='navy')
plt.plot(np.exp(equal_weight_perf.cumsum()) * 100, label='Equal Weight Baseline', linewidth=1.5, color='gray', linestyle='--')
plt.plot(np.exp(spy_benchmark.cumsum()) * 100, label='S&P 500 Benchmark (SPY)', linewidth=1.5, color='darkorange', alpha=0.7)
plt.title("Out-of-Sample Wealth Index Growth Comparison ($100 Investment)", fontsize=14, fontweight='bold')
plt.ylabel("Portfolio Value ($)")
plt.xlabel("Timeline Horizon")
plt.legend(loc="upper left")
plt.tight_layout()
plt.show()

display(metrics_df)
```
